```python
# Step 1: Define keywords for sentiment
positive_keywords = ['good', 'excellent', 'fast', 'great', 'amazing', 'like']
negative_keywords = ['bad', 'poor', 'rude', 'slow', 'worst']

# Step 2: Create lists to store feedbacks by category
all_feedbacks = []
positive_feedbacks = []
negative_feedbacks = []
neutral_feedbacks = []

# Step 3: Take input from user
num = int(input("Enter number of feedbacks: "))

for i in range(num):
    feedback = input(f"Enter feedback {i+1}: ")
    all_feedbacks.append(feedback)  # store original feedback

# Step 4: Process each feedback
for feedback in all_feedbacks:
    # Clean text: lowercase + remove extra spaces
    cleaned_feedback = feedback.lower().strip()

    # Check for positive and negative keywords
    has_positive = any(word in cleaned_feedback for word in positive_keywords)
    has_negative = any(word in cleaned_feedback for word in negative_keywords)

    # Classify using conditional statements
    if has_positive and not has_negative:
        positive_feedbacks.append(feedback)
    elif has_negative and not has_positive:
        negative_feedbacks.append(feedback)
    else:
        neutral_feedbacks.append(feedback)  # both present OR none present

# Step 5: Display results
print("\n----- Feedback Analysis Result -----")
print(f"\nTotal Feedbacks: {len(all_feedbacks)}")

print(f"\nPositive Feedbacks ({len(positive_feedbacks)}):")
for fb in positive_feedbacks:
    print(f"- {fb}")

print(f"\nNegative Feedbacks ({len(negative_feedbacks)}):")
for fb in negative_feedbacks:
    print(f"- {fb}")

print(f"\nNeutral Feedbacks ({len(neutral_feedbacks)}):")
for fb in neutral_feedbacks:
    print(f"- {fb}")

print("\n----- Insights -----")
print(f"Positive Count: {len(positive_feedbacks)}")
print(f"Negative Count: {len(negative_feedbacks)}")
print(f"Neutral Count: {len(neutral_feedbacks)}")

# Step 6: Keyword search feature
print("\n----- Search Feature -----")
search_word = input("Enter a keyword to search: ").lower()

results = []
for fb in all_feedbacks:
    if search_word in fb.lower():
        results.append(fb)

print("\nResults:")
if results:
    for res in results:
        print(f"- {res}")
else:
    print("No feedback found with that keyword.")
```
